```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("park_data/train_df.csv")
test = pd.read_csv("park_data/test_df.csv")
age_gender = pd.read_csv("park_data/age_gender_info.csv")
sub = pd.read_csv("park_data/sample_submission.csv")
```

### 컬럼명 바꿈

```python
train.columns = ['단지코드', '총세대수', '임대건물구분', '지역', '공급유형', '전용면적', '전용면적별세대수', '공가수',
       '자격유형', '임대보증금', '임대료', '지하철수',
       '버정수', '단지내주차면수', '등록차량수']

test.columns = ['단지코드', '총세대수', '임대건물구분', '지역', '공급유형', '전용면적', '전용면적별세대수', '공가수',
       '자격유형', '임대보증금', '임대료', '지하철수',
       '버정수', '단지내주차면수']
```

### 자격유형 결측값 대체

```python
test.loc[test['자격유형'].isnull() & test['단지코드'].isin(['C2411']) ,'자격유형'] = 'A'
test.loc[test['자격유형'].isnull() & test['단지코드'].isin(['C2253']) ,'자격유형'] = 'C'
```

### 지하철 결측값있는데이터 train, test로 나누기

```python
test_df = train[train.지하철수.isnull()]
train_df = train[~(train.지하철수.isnull())]
```

### 인덱스 초기화

```python
train_df.reset_index(drop=True, inplace=True)
test_df.reset_index(drop=True, inplace=True)
```

```python
train_df
```

### 임대보증금, 임대료 결측값 해결, 지하철수 type 변경

```python
train_df.loc[train_df.임대보증금=='-', '임대보증금'] = np.nan
test_df.loc[test_df.임대보증금=='-', '임대보증금'] = np.nan
test.loc[test.임대보증금=='-', '임대보증금'] = np.nan
train_df['임대보증금'] = train_df['임대보증금'].astype(float)
test_df['임대보증금'] = test_df['임대보증금'].astype(float)
test['임대보증금'] = test['임대보증금'].astype(float)

train_df.loc[train_df.임대료=='-', '임대료'] = np.nan
test_df.loc[test_df.임대료=='-', '임대료'] = np.nan
test.loc[test.임대료=='-', '임대료'] = np.nan
train_df['임대료'] = train_df['임대료'].astype(float)
test_df['임대료'] = test_df['임대료'].astype(float)
test['임대료'] = test['임대료'].astype(float)
```

```python
train_df[['임대보증금', '임대료']] = train_df[['임대보증금', '임대료']].fillna(0)
test_df[['임대보증금', '임대료']] = test_df[['임대보증금', '임대료']].fillna(0)
test[['임대보증금', '임대료']] = test[['임대보증금', '임대료']].fillna(0)
```

### 라벨인코딩

```python
train_df.자격유형.unique()
```

```python
mapping = { 'A':1, 'B':2, 'C':3, 'D':4,
            'E':5, 'F':6, 'G':7, 'H':8,
            'I':9, 'J':10, 'K':11, 'L':12, 'M':13, 'N':14, 'O':15  }
train_df['자격유형'] =train_df['자격유형'].map(mapping).astype(int)
test_df['자격유형'] =test_df['자격유형'].map(mapping).astype(int)
test['자격유형'] =test['자격유형'].map(mapping).astype(int)
```

```python
map_region = {'경상북도':1, '경상남도':2, '대전광역시':3, '경기도':4, '전라북도':5,
              '강원도':6, '광주광역시':7, '충청남도':8, '부산광역시':9, '제주특별자치도':10,
            '울산광역시':11, '충청북도':12, '전라남도':13, '대구광역시':14, '세종특별자치시':15, '서울특별시':16}
train_df["지역"] = train_df['지역'].map(map_region).astype(int)
test_df["지역"] = test_df['지역'].map(map_region).astype(int)
test["지역"] = test['지역'].map(map_region).astype(int)
```

```python
test_df['공급유형'].unique()
```

```python
map3 = {'국민임대':1, '공공임대(50년)':2, '영구임대':3, '임대상가':4, '공공임대(10년)':5,
              '공공임대(분납)':6, '행복주택':7, '공공임대(5년)':8, '장기전세':9, '공공분양':10 }
train_df["공급유형"] = train_df['공급유형'].map(map3).astype(int)
test_df["공급유형"] = test_df['공급유형'].map(map3).astype(int)
test["공급유형"] = test['공급유형'].map(map3).astype(int)
```

```python
map4 = {'아파트':1, '상가':2}
train_df["임대건물구분"] = train_df['임대건물구분'].map(map4).astype(int)
test_df["임대건물구분"] = test_df['임대건물구분'].map(map4).astype(int)
test["임대건물구분"] = test['임대건물구분'].map(map4).astype(int)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
# '총세대수', '전용면적', '전용면적별세대수','임대건물구분','지역',
# '공급유형', '공가수','자격유형', '버정수', '임대보증금','임대료',
#        '단지내주차면수'
sel = ['총세대수','임대건물구분','지역',
'공급유형', '공가수','자격유형', '버정수','단지내주차면수']

X_train = train_df[sel]
y_train = train_df['지하철수']
X_test = test_df[sel]
```

### 여러모델 테스트

```python
model = LinearRegression()
model.fit(X_train, y_train)
pred = model.predict(X_test)

print("학습(score) :", model.score(X_train, y_train) ) # 결정계수
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
pred = model.predict(X_test)

print("학습(score) :", model.score(X_train, y_train) ) # 결정계수
```

```python
model = KNeighborsClassifier(n_neighbors=1)
model.fit(X_train, y_train)
pred = model.predict(X_test)

print("학습(score) :", model.score(X_train, y_train) ) # 결정계수
```

```python
model = DecisionTreeClassifier(max_depth=7, random_state=0)
model.fit(X_train, y_train)
pred = model.predict(X_test)

print("학습(score) :", model.score(X_train, y_train) ) # 결정계수
```

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
pred = model.predict(X_test)

print("학습(score) :", model.score(X_train, y_train) ) # 결정계수
```

```python
model = GradientBoostingClassifier(random_state=0, max_depth=1, learning_rate=0.01)
model.fit(X_train, y_train)
pred = model.predict(X_test)

print("학습(score) :", model.score(X_train, y_train) ) # 결정계수
```
